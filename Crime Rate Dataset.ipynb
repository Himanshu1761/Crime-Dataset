# Comapring Last three years IPC Crimes -2019,2021 & 2021 

```python
data=pd.DataFrame(data)
df={
    'State-Name':data['State/UT'],
    'IPC Crimes-2019':data['IPC Crimes-2019'],
    'IPC Crimes-2020':data['IPC Crimes-2020'],
    'IPC Crimes-2021':data['IPC Crimes-2021']
}
df=pd.DataFrame(df)


plt.figure(figsize=(15,60))
t=1
for index, row in df.iterrows():
    labels = ['IPC Crimes-2019', 'IPC Crimes-2020', 'IPC Crimes-2021']
    sizes = [row['IPC Crimes-2019'], row['IPC Crimes-2020'], row['IPC Crimes-2021']]
    plt.subplot(15,3,t)
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#E3735E','beige','#89CFF0'])
    plt.title(f"IPC Crime for {row['State-Name']}")
    t+=1
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.tight_layout()
plt.show()
```

# Comapring Last three years SLL Crimes -2019,2021 & 2021 

```python
df={
    'State-Name':data['State/UT'],
    'SLL Crimes-2019':data['SLL Crimes-2019'],
    'SLL Crimes-2020':data['SLL Crimes-2020'],
    'SLL Crimes-2021':data['SLL Crimes-2021']
}

df=pd.DataFrame(df)

t=1
plt.figure(figsize=(15,60))
for i,j in df.iterrows():
    labels=['SLL Crimes-2019','SLL Crimes-2020','SLL Crimes-2021']
    sizes=[j['SLL Crimes-2019'],j['SLL Crimes-2020'],j['SLL Crimes-2021']]
    
    
    plt.subplot(15,3,t)
    plt.pie(sizes, labels=labels, autopct='%.2f',startangle=140, colors=['#E3735E','beige','#89CFF0'])
    plt.title(f"SLL Crimes in {j['State-Name']}")
    t+=1
    
plt.tight_layout()
plt.show()
```

# Mapping the States into different Regions

Northern Zonal-(Chandigarh, Delhi, Haryana, Himachal Pradesh, Jammu and Kashmir, Ladakh, Punjab, Rajasthan)
North Eastern- (Assam, Arunachal Pradesh, Manipur, Meghalaya, Mizoram, Nagaland, Tripura, Sikkim)
Central Zonal- (Chhattisgarh, Madhya Pradesh, Uttarakhand , Uttar Pradesh)
Eastern Zonal-(Bihar, Jharkhand, Odisha, West Bengal)
Western Zonal- (Dadra and Nagar Haveli, Daman and Diu, Goa, Gujarat, Maharashtra)
Southern Zonal-(Andhra Pradesh, Karnataka, Kerala, Puducherry, Tamil Nadu, Telangana)

```python
region = {
    ('Chandigarh', 'Delhi', 'Haryana', 'Himachal Pradesh', 'Jammu and Kashmir', 'Ladakh', 'Punjab', 'Rajasthan'): 'Northern',
    ('Assam', 'Arunachal Pradesh', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Tripura', 'Sikkim'): 'North Eastern',
    ('Chhattisgarh', 'Madhya Pradesh', 'Uttarakhand' ,'Uttar Pradesh'): 'Central Zonal',
    ('Bihar', 'Jharkhand', 'Odisha','West Bengal'): 'Eastern Zonal',
    ('Dadra and Nagar Haveli and Daman and Diu', 'Goa', 'Gujarat', 'Maharashtra'): 'Western Zonal',
    ('Andhra Pradesh', 'Karnataka', 'Kerala', 'Puducherry', 'Tamil Nadu', 'Telangana','Andaman and Nicobar Islands','Lakshadweep'): 'Southern Zonal'
    
}

for states, region_label in region.items():
    data.loc[data['State/UT'].isin(states), 'Region'] = region_label
    
```

```python
region_mapping = {
    'Total States': 'Total States'
}

# Replace specific state names with region labels
data['Region'] = data['State/UT'].replace(region_mapping)
```
